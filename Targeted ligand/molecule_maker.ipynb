```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem  
from rdkit.Chem import PandasTools
import pubchempy as pcp

from rdkit.Avalon import pyAvalonTools
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem import MACCSkeys
from rdkit import DataStructs

import pandas as pd
import numpy as np
import json
from tqdm import tqdm



#1.1 function for getting molecule from name

def get_molecule_from_name(name):
    # Search for the compound by name
    compounds = pcp.get_compounds(name, 'name')
    if compounds:
        print(compounds[0].to_dict())
        smiles = compounds[0].isomeric_smiles
        molecule = Chem.MolFromSmiles(smiles)
        return smiles,molecule
    else:
        return None
    
# generating sdf molecule from smiles
def generate_sdf_from_smiles(smiles,filename):
    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol)
    AllChem.MMFFOptimizeMolecule(mol)
    writer = Chem.SDWriter(filename)
    writer.write(mol)
    writer.close()
    return mol
# Example usage
smiles= 'CC(=O)[C@H]1CC[C@@H]2[C@@]1(CC[C@H]3[C@H]2CCC4=CC(=O)CC[C@]34C)C'
filename='STR.sdf'
EST_smile = 'C[C@]12CC[C@H]3[C@H]([C@@H]1CC[C@@H]2O)CCC4=C3C=CC(=C4)O'
filename_est='EST.sdf'
AND_smiles = 'C[C@]12CC[C@H]3[C@H]([C@@H]1CCC2=O)CC=C4[C@@]3(CC[C@@H](C4)O)C'
filename_and='AND.sdf'
Tes_smile = 'C[C@]12CC[C@H]3[C@H]([C@@H]1CC[C@@H]2O)CCC4=CC(=O)CC[C@]34C'
filename_tes='TES.sdf'
mol=generate_sdf_from_smiles(EST_smile,filename_est)
print(Chem.MolToSmiles(mol))
```
